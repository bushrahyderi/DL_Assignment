```python
import numpy as np

class gates():
    def __init__(self, not_weight=np.array([1]), and_weight=np.array([-1, 1]),
                 or_weight=np.array([1, -4]), or_biase=1, and_biase=-7, not_biase=0.1,
                 initial_learning_rate=0.01) -> None:
        self.not_weight = not_weight
        self.and_weight = and_weight
        self.or_weight = or_weight
        self.or_biase = or_biase
        self.and_biase = and_biase
        self.not_biase = not_biase
        self.learning_rate = initial_learning_rate

    def activation(self, x):
        return 1 if x >= 0 else 0
    
    def perceptron(self, x, w, b):
        res = np.dot(x, w) + b
        return self.activation(res)
    
    def NOT_logicFunction(self, x):
        x = np.array(x)
        return self.perceptron(x, self.not_weight, self.not_biase)
    
    def AND_logicFunction(self, x):
        x = np.array(x)
        return self.perceptron(x, self.and_weight, self.and_biase)
    
    def OR_logicFunction(self, x):
        x = np.array(x)
        return self.perceptron(x, self.or_weight, self.or_biase)
    
    def XOR_logicFunction(self, x):
        x = np.array(x)

        
        y0 = self.NOT_logicFunction(x[0])
        y1 = self.NOT_logicFunction(x[1])
        z1 = self.AND_logicFunction([y0, x[1]])
        z2 = self.AND_logicFunction([y1, x[0]])
        
        return self.OR_logicFunction([z1, z2])

    def adaptive_learning_rate(self, error, decay_rate=0.01):   
       return self.learning_rate / (1 + decay_rate * abs(error))

    def update_weight(self, inp, out, tar, weight, max_clip=1.0):
        error = tar - out
        adaptive_lr = self.adaptive_learning_rate(error)  # Adaptive learning rate
        update = adaptive_lr * error * inp
        new_weight = weight + update
        
        # Apply clipping to the weights
        new_weight = np.clip(new_weight, -max_clip, max_clip) #modification
        
        return new_weight

    def update_not(self):
        inp = np.array([1, 0])
        tar = np.array([0, 1])
        
        i = 0
        while i < len(inp):
            out = self.NOT_logicFunction(inp[i])
            if tar[i] != out:
                weight = self.not_weight
                self.not_weight = self.update_weight(inp[i], out, tar[i], self.not_weight)
                self.not_biase += 0.1 * (tar[i] - out)  # Keep some bias update if needed
                print(self.not_weight, self.not_biase)
                i = 0
            else:
                i += 1
                
    def update_and(self):
        inp = np.array([(0, 0), (1, 0), (0, 1), (1, 1)])
        tar = np.array([0, 0, 0, 1])
        
        i = 0
        while i < len(inp):
            out = self.AND_logicFunction(inp[i])
            if tar[i] != out:
                weight = self.and_weight
                self.and_weight = self.update_weight(inp[i], out, tar[i], self.and_weight)
                self.and_biase += 0.1 * (tar[i] - out)  # Keep some bias update if needed
                print(self.and_weight, self.and_biase)
                i = 0
            else:
                i += 1
                
    def update_or(self):
        inp = np.array([(0, 0), (1, 0), (0, 1), (1, 1)])
        tar = np.array([0, 1, 1, 1])
        
        i = 0
        while i < len(inp):
            out = self.OR_logicFunction(inp[i])
            if tar[i] != out:
                weight = self.or_weight
                self.or_weight = self.update_weight(inp[i], out, tar[i], self.or_weight)
                self.or_biase += 0.1 * (tar[i] - out)  # Keep some bias update if needed
                print(self.or_weight, self.or_biase)
                i = 0
            else:
                i += 1

test1 = np.array([0, 1])
test2 = np.array([1, 1])
test3 = np.array([0, 0])
test4 = np.array([1, 0])
output = [1, 0, 0, 1]

test_data = [test1, test2, test3, test4]

perceptron = gates(
    not_weight = 2, and_weight = np.array([0, 1]), or_weight = np.array([1, 1]), initial_learning_rate=0.3, not_biase= 0.5, and_biase = -1.5, or_biase = -0.5
)

perceptron.update_and()
perceptron.update_or()
perceptron.update_not()

for i in range(len(test_data)):
    result = perceptron.XOR_logicFunction(test_data[i])
    print("Predicted_value: ", result)

print("")
print("Weights after training")
print("OR weight:", perceptron.or_weight)
print("AND weight:", perceptron.and_weight)
print("NOT weight:", perceptron.not_weight)
print("")
```
